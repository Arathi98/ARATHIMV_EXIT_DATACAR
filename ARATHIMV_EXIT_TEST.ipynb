TREATING NULL VALUES

```python
df['Engine Fuel Type'].fillna(df['Engine Fuel Type'].mode()[0], inplace = True)
df['Engine HP'].fillna(df['Engine HP'].mean(), inplace = True)
df['Engine Cylinders'].fillna(df['Engine Cylinders'].mode()[0], inplace = True)
df['Number of Doors'].fillna(df['Number of Doors'].mode()[0], inplace = True)
df['Market Category'].fillna(df['Market Category'].mode()[0], inplace = True)
df.isnull().sum()
```

```python
cat_cols = df.select_dtypes(include=['object']).columns
num_cols = df.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

FINDING OUTLIERS

```python
for col in num_cols:
    print(col)
    print('Skew :', round(df[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    df[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.show()
```

TREATING OUTLIERS

```python
for i in ['Year', 'Engine HP', 'Engine Cylinders','highway MPG', 'city mpg','MSRP']:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR=Q3-Q1
    upper_limit=Q3+1.5*IQR
    lower_limit=Q1-1.5*IQR
    df[i].where(df[i] > lower_limit, lower_limit, inplace = True)
    df[i].where(df[i] < upper_limit, upper_limit, inplace = True)
```

SPLLITING INTO X AND Y

```python
X = df.drop(['MSRP'], axis=1)
y = df['MSRP']
```

ENCODING AND SCALING

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
transformer = make_column_transformer(
    (OneHotEncoder(drop='first', sparse_output=False),['Make', 'Model', 'Engine Fuel Type', 'Transmission Type',
     'Driven_Wheels', 'Market Category', 'Vehicle Size', 'Vehicle Style']),
    (StandardScaler(),['Engine HP', 'Engine Cylinders', 'Number of Doors', 'highway MPG', 'city mpg', 'Popularity'] ),
    remainder='passthrough')

transformed = transformer.fit_transform(X)
transformed.shape
```

```python
transformed_df = pd.DataFrame(transformed, columns=transformer.get_feature_names_out(transformer.feature_names_in_))
transformed_df.head()
```

```python
transformed_df.dtypes
```

TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(transformed_df, y, test_size=0.2, random_state=42)
```

MODELING

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python


# Import different regression algorithms
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

# Create a list of regressors to try
regressors = [
    LinearRegression(),
    DecisionTreeRegressor(),
    RandomForestRegressor(),
    GradientBoostingRegressor()
]

# Train and evaluate different regressors
for regressor in regressors:
    model = regressor  # Choose a regressor from the list

    # Train the model
    model.fit(X_train, y_train)

    # Make predictions on the testing set
    y_pred = model.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print("Regressor:", type(regressor).__name__)
    print("Mean Squared Error:", mse)
    print("Mean Absolute Error:", mae)
    print("R-squared Score:", r2)
    print("---------------------")

```

Among these three models, the Random Forest Regressor exhibits the lowest Mean Squared Error (9782484.34) 
and Mean Absolute Error (1996.68), indicating better predictive accuracy and smaller deviations between the predicted
and actual car prices. Additionally, it also has a high R-squared Score (0.975), suggesting a better fit of the model
to the data compared to the other models.

Random forest is the best model to predict price of cars here

FINETUNING

```python


from sklearn.model_selection import GridSearchCV

# Define hyperparameters to tune for each regressor
# Modify the parameter grids based on the specific regressors and parameters you want to tune
param_grid = {
    'LinearRegression': {},
    'DecisionTreeRegressor': {'max_depth': [None, 5, 10]},
    'RandomForestRegressor': {'n_estimators': [100, 200, 300], 'max_depth': [None, 5, 10]},
    'GradientBoostingRegressor': {'learning_rate': [0.1, 0.01], 'n_estimators': [100, 200]}
}

best_regressor = None
best_score = 0

# Perform hyperparameter tuning and cross-validation for each regressor
for regressor in regressors:
    model = regressor

    # Define the parameter grid for the current regressor
    params = param_grid[type(model).__name__]

    # Perform grid search with cross-validation
    grid_search = GridSearchCV(model, params, scoring='neg_mean_squared_error', cv=5)
    grid_search.fit(X_train, y_train)

    # Get the best parameters and score
    best_params = grid_search.best_params_
    best_score = grid_search.best_score_

    print("Regressor:", type(regressor).__name__)
    print("Best Parameters:", best_params)
    print("Best Score (negative mean squared error):", best_score)
    print("---------------------")

    # Update the best regressor and score if necessary
    if best_score > best_score:
        best_regressor = grid_search.best_estimator_

# Use the best-performing regressor for final evaluation
if best_regressor is not None:
    # Make predictions on the testing set
    y_pred = best_regressor.predict(X_test)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print("Best Regressor:", type(best_regressor).__name__)
    print("Mean Squared Error:", mse)
    print("Mean Absolute Error:", mae)
    print("R-squared Score:", r2)
    print("---------------------")

    
```
